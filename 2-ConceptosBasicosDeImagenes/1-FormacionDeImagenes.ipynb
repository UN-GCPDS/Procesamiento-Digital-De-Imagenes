## Espectro de Luz Visible 

El fenómeno físico más usado para la generación de imágenes es la luz. Lo que denominamos luz es el espectro electromagnético visible. Las cámaras digitales, que son usadas ampliamente, son sensores especializados para la captura de este espectro (igual que el ojo humano). Este espectro visible comprende de los 380 nm a los 760 nm, como se muestra en la figura.

Como se dijo, la luz no es el único fenómeno usando para la generación de imágenes. También se usan otras longitudes de onda, como por ejemplo el infrarrojo (900 nm a 14000 nm) para capturar temperaturas, o fenómenos completamente diferentes como el sonido, el cual es usado para la generación de ecografías.

## Modelos de Cámara

La cámara intenta simular el funcionamiento del ojo humano

### Ojo humano

Después del cerebro, los ojo el sistema más complejo del cuerpo humano. Es el encargado de transformar la luz en señales eléctricas que luego será procesadas por el cerebro.

#### Células Fotorreceptoras 

Ubicadas en la retina, existen células fotorreceptoras encargadas de transformar la luz en impulsos eléctricos. Hay dos tipos; los conos y bastones. 

**Bastones**
* Responsables de la visión en bajas condiciones de luminosidad
* No son capaces de percibir el color
* Encargados de la visión periférica (Captura de movimiento)
* Su concentración es más alta en los extremos de la retina 
* Aproximadamente hay 130 millones de bastones en el ojo humano 

**Conos** 
* Encargados de percibir los colores 
* Especializados en capturar los colores rojo, verde y azul
* Su concentración es más alta en la parte interna de la retina
* Aproximadamente hay 7 millones de conos en el ojo humano

### Cámara Pinhole 

El modelo de cámara pinhole es el más sencillo. La idea principal es evitar que múltiples rayos de luz que proviene del mismo punto incidan en toda la superficie de captura. Por ello se usa una barrera y en ella un pequeño agujero. Los principales problemas de esta aproximación es que para generar una imagen nítida se requiere un agujero muy pequeño que a la vez permite pasar poca luz resultando en imágenes con poco brillo.

#### Sistema Coordenado del Mundo Físico al de la Cámara 

Para transformar los puntos del sistema coordenado físico al de la cámara usamos una matriz de rotación $R \in \mathbb{R}^{3 \times 3}$ y un vector de traslación $t \in \mathbb{R}^3$

$$
\begin{bmatrix}
X_c \\ Y_c \\ Z_c
\end{bmatrix} = R \begin{bmatrix}
X_w \\ Y_w \\ Z_w
\end{bmatrix} + t 
$$


$$
\begin{bmatrix}
X_c \\ Y_c \\ Z_c
\end{bmatrix} = [R|t] \begin{bmatrix}
X_w \\ Y_w \\ Z_w \\ 1
\end{bmatrix} 
$$

Los parámetros **extrinsecos** de la cámara son $P = [R|t]$.

#### Sistema Coordenado de la Cámara al de la Imagen 


Por similaridad de triángulos podemos escribir

$$x = f\frac{X_c}{Z_c}$$
$$y = f\frac{Y_c}{Z_c}$$


$$ \begin{bmatrix}
x' \\ y' \\ z'
\end{bmatrix} = \begin{bmatrix} f & 0 & 0 \\ 0 & f & 0 \\ 0 & 0 & 1 \\
\end{bmatrix} \begin{bmatrix}
X_c \\ Y_c \\ Z_c
\end{bmatrix} $$




Los parámetros **intrínsecos** de la cámara son $K = \begin{bmatrix} f & 0 & 0 \\ 0 & f & 0 \\ 0 & 0 & 1 \\ \end{bmatrix}$ en un caso ideal. Pero por la no perfección del lente, pueden existir diferentes distancias focales para cada eje $f_x$ y $f_y$. Además, el centro focal tampoco coinciden con el origen de la imagen. Por último ademas $\gamma$



$$K = \begin{bmatrix} f_x & \gamma & c_x \\
0 & f_y & c_y \\
0 & 0 & 1 \\
\end{bmatrix}$$


$$ \begin{bmatrix}
u' \\ v' \\ w'
\end{bmatrix} =  \begin{bmatrix} f_x & \gamma & c_x \\
0 & f_y & c_y \\
0 & 0 & 1 \\
\end{bmatrix}
\begin{bmatrix}
X_c \\ Y_c \\ Z_c
\end{bmatrix} $$


$$u =\frac{u'}{w'}$$

$$v =\frac{v'}{w'}$$

## Cámara Digital: Filtro Bayer

Simulando el ojo humano, usamos el filtro de Bayer para obtener color en las imágenes, cabe mencionar que hay diferentes arreglos de sensores, pero este es el más común. El principio es el mismo del ojo, hacer que los sensores solo respondan las longitudes de onda correspondientes al rojo, verde y azul.  

<center>
<figure>
<img src='https://static-01.hindawi.com/articles/vlsi/volume-2013/738057/figures/738057.fig.001.jpg' width="400"> 
<figcaption><a href="https://www.hindawi.com/journals/vlsi/2013/738057/">Filtro Bayer</a></figcaption>
</figure>
</center>


Entonces para representar el color usamos combinaciones de tres colores, rojo, verde y azul (Red, Blue and Green). Del uso de un conjunto finito de colores debemos el concepto de **canales**, donde un canal representa un color. En términos espaciales, el sensor generalmente (depende de la compañía fabricante) combina grupos de cuatro fotosensores para formar la información de color de un pixel. Un pixel es la unidad básica de una imagen.
